### <font color='black' face='微软雅黑'>1.2. 函数参数</font><a name='1.2'></a>
[<font color='black' face='微软雅黑'>回到章首</font>](#1)

将上面 `iplot()` 函数签名的信息重新整理在下图中，故意用浅灰色淡化不重要的参数：

<img src='iplot function.png'>

绘图函数 `iplot()` 一招鲜吃遍天，但细节在魔鬼，了解其参数越细就能实现越精巧的图。下面事无巨细的解释该函数中的常用参数，大体分为三类：

- **<font color='red' face='微软雅黑'>通用参数</font>**：所有类型图都有的特征，比如 `kind`、`title`、`theme` 等
- **<font color='blue' face='微软雅黑'>条件参数</font>**：在某些条件下才被“激活“的特征，比如 `symbol` 和 `size` 仅当参数 `mode='marker'` 才使用
- **<font color='green' face='微软雅黑'>额外参数</font>**：用于进一步个性化的参数 `**kwargs` 

### <font color='red' face='微软雅黑'>通用参数</font>

    kind：字符串格式，用于设置图的种类，具体值包括

        散点图 scatter、柱状图 bar、箱形图 box、差异图 spread、比率图 ratio、热力图 heatmap、
        平面图 surface、直方图 histogram、气泡图 bubble、三维气泡图 bubble3d、三维散点图 scatter3d、
        地理散点图 scattergeo、K 线图 ohlc、蜡烛图 candle、饼状图 pie、地图 choropleth

    title：字符串格式，用于设置图表标题

        xTitle：字符串格式，用于设置 x 轴标题
        yTitle：字符串格式，用于设置 y 轴标题

    x：字符串格式，数据帧中用于 x 轴变量的列标签
    y：字符串格式，数据帧中用于 y 轴变量的列标签

    theme：字符串格式，用于设置主题风格，可用 cf.getThemes() 查看，具体值包括 ggplot、pearl、solar、space、 white、polar、
    henanigans。

```python
cf.getThemes()
```

```python
for theme in cf.getThemes():
    pd.DataFrame( data=np.sin(np.linspace(0,2*np.pi,100)), ).iplot(theme=theme,  title=f'{theme} theme')
```

    fill：布尔格式，用于填充轨迹 (trace) 和坐标轴之间的空白。

    colors：字典、列表或字符串格式，用于设置颜色

        字典：{column:color} 按数据帧中的列标签设置颜色
        列表：[color] 对每条轨迹按顺序的设置颜色
        字符串：具体颜色的英文名称，适用于所有轨迹

    colorscale：字符串格式，用于设置色阶，具体名称可用 cf.colors.scales() 查看，如下图所示。在颜色字符串前加 – 前缀可以反转色阶。
    举例 “accent” 和 “-accent” 的色阶顺序正好相反。

```python
cf.colors.scales()
```

    width：字典、列表或整数格式，用于设置轨迹宽度

        字典：{column:value} 按数据帧中的列标签设置宽度
        列表：[value] 对每条轨迹按顺序的设置宽度
        整数：具体数值，适用于所有轨迹

    dash：字典、列表或字符串格式，用于设置轨迹风格

        字典：{column:value} 按数据帧中的列标签设置风格
        列表：[value] 对每条轨迹按顺序的设置风格
        字符串：具体风格的名称，适用于所有轨迹

    具体选项有实线 solid、虚线 dash、虚点 dashdot、点 dot。

    mode：字典、列表或字符串格式，用于设置轨迹模式

        字典：{column:value} 按数据帧中的列标签设置模式
        列表：[value] 对每条轨迹按顺序的设置模式
        字符串：具体模式的名称，适用于所有轨迹

    具体选项有折线 lines、散点 markers、折线加散点 lines+markers、折线加文字 lines+text、散点加文字 markers+text、折线加散点和文字 
    lines+markers+text。

    interpolation：字典、列表或字符串格式，用于设置插值方法

        字典：{column:value} 按数据帧中的列标签设置插值方法
        列表：[value] 对每条轨迹按顺序的设置插值方法
        字符串：具体插值方法的名称，适用于所有轨迹

    具体选项有线性 linear、三次样条 spline、 阶梯 hv 或 vh 或 vhv 或 hvh。

    annotations：字典格式 {x_point: text}，用于在点 x_point上标注 text。

    keys：列表格式，指定数据帧中的一组列标签用于排序。

    bestfit：布尔或列表格式，用于拟合数据。

        布尔：True 对所有列的数据都做拟合
        列表：[columns] 对列表中包含列的数据做拟合

    bestfit_colors：字典或列表格式，用于设定数据拟合线的颜色。

        字典：{column:color} 按数据帧中的列标签设置颜色
        列表：[color] 对每条轨迹按顺序的设置颜色

    categories：字符串格式，数据帧中用于区分类别的列标签

    text：字符串格式，数据帧用于显示文字的列标签
 
    gridcolor：字符串格式，用于设定网格颜色
    zerolinecolor：字符串格式，用于设定零线颜色

    secondary_y：字符串格式，数据帧中用于第二个 y 轴变量的列标签
    secondary_y_title：字符串格式，用于设置第二个 y 轴标题

    subplots：布尔格式，如果 True 则画子图

    error_x：整数或浮点数格式，用于设置 x 轴变量的误差值
    error_y：整数或浮点数格式，用于设置 y 轴变量的误差值

    error_type：字符串格式，用于设定误差条形的类型，具体选项有 data、constant、percent、sqrt、continuous、continuous_percent。

最后将图存成不同格式的布尔型参数：

    asFigure：如果 True 则将图存成 Plotly 图格式
    asImage：如果 True 则将图存成 PNG 格式
    asPlot：如果 True 则将图在浏览器打开
    asUrl：如果 True 则将返回网址 (online 模式) 或返回本地路径 (offline 模式)

### <font color='blue' face='微软雅黑'>条件参数</font>

    (1)	当 subplots = True 才适用，即多图

        shape：元组格式 (rows, cols)，其中 rows 是每行包含子图的个数，cols 是每列包含子图的个数。

    (2)	当 3D 图才使用

        zTitle：字符串格式，用于设置 z 轴标题

        z：字符串格式，数据帧中用于 z轴变量的列标签

    (3)	当 mode 含 marker 才适用

        symbol：字典、列表或字符串格式，用于设置标记类型

            字典：{column:value} 按数据帧中的列标签设置标记类型
            列表：[value] 对每条轨迹按顺序的设置标记类型
            字符串：具体标记类型的名称，适用于所有轨迹

        具体选项有圆形 circle、圆形加点 circle-dot、菱形 diamond、方形 square 等。

        size：字符串或整数格式，用于设置标记大小

    (4)	当 kind = bar 才适用，即条形图

        barmode：字符串格式，用于设置柱状类型，具体选项有分组 group、堆叠 stack、重叠 overlay。

        sortbars：布尔格式，用于递减排列柱状。

    (5)	当 kind = histogram 才适用，即直方图

        bins：整数或元组格式

            整数：设置桶个数
            元组：(start, end, size) 设置起始点、终止点、和桶的大小

        histnorm：字符串格式。具体选项有

            频率 frequency (桶高等于计数)
            百分比 percent (桶高等于每桶中计数占总数的比例)
            概率 probability (桶高等于概率，桶宽为 1)
            密度 density (桶高等于计数除以桶宽，使得桶面积加起来等于桶的总数)
            概率密度 probability density (桶高等于概率除以桶宽，使得桶面积加起来等于 1)

        histfunc：字符串格式，用于设置桶函数。具体选项有计数函数 count、求和函数 sum、平均函数 avg、最小值函数 min、最大值函数 max。

    (6)	当 kind = box 才适用，即箱形图

        boxpoints：布尔或字符串格式，用于在箱形图中显示数据，仅当 kind = box 才适用，具体选项有离群值 outliers、全部 all、可疑离群
        值 suspectedoutliers、不显示 False。

    (7)	当 kind = pie 才适用，即饼状图

        labels：字符串格式，将数据帧中的里列标签设为饼状图每块的标签。

        values：字符串格式，将数据帧中的列数据的值设为饼状图每块的面积。

    (8)	当 kind = bar 或 histogram才适用，即条形图或直方图

        bargap：浮点数格式，值在 0 和 1 之间，用于设置柱状的间隔。

        bargroupgap：浮点数格式，值在 0 和 1 之间，用于设置柱状分组的间隔。

    (9)	当 kind = bar 或 histogram 或 box 才适用，即条形图、直方图或箱形图

        orientation：字符串格式，用于设置形状的排放方式，h 代表水平 v 代表竖直。

### <font color='green' face='微软雅黑'>额外参数</font>

    (1)	用于散点图 (scatter)

        connectgaps：布尔格式，如果为 True，那么空值的点被连接起来。

    (2)	用于直方图 (historgram)

        linecolor：字符串格式，设置每个桶的颜色。

    (3)	用于饼状图 (pie)

        sort：布尔格式，如果为 True，根据值大小对标签排序。

        pull：浮点数格式，值在 0 和 1 之间，设定从中心往外拉出块的程度。

        hole：浮点数格式，值在 0 和 1 之间，设定内饼的半径长度占比。

        linecolor：字符串格式，设置每块等值线的颜色。

        linewidth：字符串格式，设置每块等值线的宽度。

        textcolor：字符串格式，设置每块里文字的颜色

        textposition：字符串格式，设置每块里文字的位置，具体值为 outside 和 inner。

        textinfo：字符串格式，设置每块里提示框的内容，具体值为

            label 表示标签
            value 表示值
            percent 表示每个值占总值得百分比

        或者它们的组合，用 '+' 连接，比如 'label+percent'

    (4)	用于热力图 (heatmap) 和平面图 (surface)

        center_scale：浮点数格式，用于设定色阶中心值，并自动设定了 zmin 和 zmax 值。

        zmin：浮点数格式，用于设定色阶最小值，会影响色阶范围。
        zmax：浮点数格式，用于设定色阶最大值，会影响色阶范围。

    (5)	用于误差棒 (error bar)

        error_trace：字符串格式，按数据帧中的列标签显示误差，如果为空则误差适用于所有轨迹。

        error_color：字符串格式，用于设置误差棒的颜色。

        error_thickness：浮点数格式，用于设置误差棒的厚度。

        error_width：浮点数格式，用于设置误差棒的宽度。

        error_opacity，值在 0 和 1 之间，用于设置误差棒的透明度。

    (6)	用于子图 (subplots)

        horizontal_spacing：浮点数格式，值在 0 和 1 之间，用于设定列之间的空白大小。

        vertical_spacing：浮点数格式，值在 0 和 1 之间，用于设定行之间的空白大小。

        shared_xaxes：布尔格式，如果为 True，每列子图会共享一个 x 轴，此 x 轴在网格最下边。

        shared_yaxes：布尔格式，如果为 True，每列子图会共享一个 y 轴，此 y 轴在网格最左边。

        subplot_titles：布尔格式，如果为 True，每个子图上都有标题。

    (7)	用于形状 (shapes)

        hline：浮点数、列表或字典格式，用于绘制水平线

            浮点数：value，y 轴单个位置
            列表：[value]，y 轴多个位置
            字典：{attribute:value}，额外修饰水平线的参数

        vline : 浮点数、列表或字典格式，用于绘制竖直线

            浮点数：value，x 轴单个位置
            列表：[value]，x 轴多个位置
            字典：{attribute:value}，额外修饰竖直线的参数

        hpsan：元组、列表或字典格式，用于绘制水平块

            元组：(y0, y1)，单块的两条边在 y 轴位置
            列表：[(y0, y1)]，多块的两条边在 y 轴位置
            字典：{attribute:value}，额外修饰水平块的参数

        vpsan：元组、列表或字典格式，用于绘制竖直块

            元组：(x0, x1)，单块的两条边在 x 轴位置
            列表：[ (x0, x1) ]，多块的两条边在 x 轴位置
            字典：{attribute:value}，额外修饰竖直块的参数

        shapes：字典和列表格式，用于绘制形状，细节信息可用 cf.tools.get_shape 查看。

            字典：{attribute:value}，形状的参数
            列表：[ {attribute:value} ]，多个形状的参数

    (8)	用于轴范围 (axis ranges)

        xrange：列表格式，[lb, ub]，用于设置 x 轴的范围
        yrange：列表格式，[lb, ub]，用于设置 y 轴的范围
        zrange：列表格式，[lb, ub]，用于设置 z 轴的范围

    (9)	范围选择器 (range selector)

        rangeselector：字典格式，创建范围选择器对象，细节信息可用 cf.tools.get_range_selector 查看。

    (10)	范围滑块 (range slider)        

        rangeslider：布尔或字典格式，创建范围滑块对象。如果为布尔 True 值，范围滑块可见；如果为字典，则用于设置范围滑块的属性。其属性
        包含 autorange、bgclor、bordercolor、borderwidth、range、thickness、visible 等。

    (11)	标注 (annotations)

        fontcolor：字符串格式，用于设定标注文字颜色

本节使用 FIFA 18 数据，包括梅西、C 罗等 18000 名球员的 45 个特征，如年龄、国籍、评分、俱乐部、市值、工资、传球、抢断、终结能力等。从 csv 读取数据并删除对数据分析无用的特征。

```python
FIFA = pd.read_csv('FIFA18.csv')
df = FIFA.drop(['Photo', 'Flag', 'Club Logo'], axis=1)
FIFA.head(3).append(FIFA.tail(3))
```

### <font color='black' face='微软雅黑'>2.1. Choropleth 图</font><a name='2.1'></a>
[<font color='black' face='微软雅黑'>回到章首</font>](#2)

`Cufflinks` 只用一个 `iplot()` 函数便可以实现高质量的可视化，生成的图基本体现了 choropleth 的特征。

    go.Scatter( x=df.rank, y=df.overall )           # plotly            matplotlib
    px.scatter( data=df, x='rank', y='overall' )    # plotly express    Seaborn 轴级函数
    df.iplot( kind='scatter', x='rank', y='overall' ) # cufflinks       Seaborn 图级函数

```python
grp = FIFA.groupby("Nationality").size().reset_index(name="Count")

grp.iplot( kind='choropleth', locations='Nationality', z ='Count',
           text = 'Nationality', locationmode = 'country names', theme='ggplot',
           colorscale='blues', title = "FIFA 18 球员国籍",
           projection = dict(type = 'natural earth') )
```

### <font color='black' face='微软雅黑'>2.2. 双 y 轴折线图</font><a name='2.2'></a>
[<font color='black' face='微软雅黑'>回到章首</font>](#2)

```python
df_ = df.nlargest(100, 'Overall')
df_['Rank'] = np.arange(1, len(df_) + 1)
```

首先用 YahooFinancials API 来下载四个股票的 2020 年到 2021 年的最新数据，安装该 API 用一行代码:

    pip install yahoofinancials

数据的描述如下

- 起始日：2020-01-01
- 终止日：2021-01-31
- 四只股票：FUTU、NIO、FUBO，DAO

下面代码就是从 API 获取数据：

```python
from yahoofinancials import YahooFinancials
```

```python
start_date = '2020-01-01'
end_date = '2021-01-31'

stock_code = ['FUTU', 'NIO', 'FUBO', 'DAO']
stock = YahooFinancials( stock_code )
stock_daily = stock.get_historical_price_data( start_date, end_date, 'daily' )
```

```python
stock_daily['FUTU']['prices']
```

该 API 返回结果是字典格式，样子非常丑陋，下面代码将上面的原始数据转换成数据帧。

```python
def data_converter( price_data, code ):
    columns = ['open', 'close', 'low', 'high' ]
    price_dict = price_data[code]['prices']
    index = [ p['formatted_date'] for p in price_dict ]
    price = [ [p[c] for c in columns] for p in price_dict ]

    data = pd.DataFrame( price,
                         index=pd.Index(index, name='date'),
                         columns=pd.Index(columns, name='OHLC') )
    return data
```

```python
df = pd.DataFrame()

for code in stock_code:
    data = data_converter( stock_daily, code )
    df[code] = data['close']

df.index = pd.to_datetime(df.index)
df.head(3).append(df.tail(3))
```

上三图的子图布局很规范，分别是 2×2、4×1 和 1×4，此外子图类型也很一致，都是折线图。

如果它们的布局不规范而且类型也不一致，那该如何画出来。`Cufflinks` 可以实现上面需求，分四步

1. **创建多图** - 将具体的图一个个按顺序放入 `cf.figures()` 里，记做 `f`
2. **创建布局** - 用内置函数 `cf.tools.get_base_layout(f)`，记做 `layout`
3. **设定位置** - 用 `cf.subplots(f, …, layout)` 并设置参数 `specs`，记做 `sp`
4. **展示图形** - 用 `cf.iplot(sp)` 画图

```python
figs = cf.figures( df.pct_change(), [dict(kind='histogram', barmode='stack')], asList=True )
figs.append( df[['FUTU','NIO']].pct_change().figure(kind='scatter', mode='markers', size=3) )
figs.append( df.pct_change().corr().figure(kind='heatmap', colorscale='pubu') )
figs.append( df['DAO'].figure(kind='line', bestfit=True, colors=['blue'], bestfit_colors=['pink']))
len(figs)
```

```python
figs = cf.figures( df.pct_change(), [dict(kind='histogram', barmode='stack')], asList=True )
figs.append( df[['FUTU','NIO']].pct_change().figure(kind='scatter', mode='markers', size=3) )
figs.append( df.pct_change().corr().figure(kind='heatmap', colorscale='pubu') )
figs.append( df['DAO'].figure(kind='line', bestfit=True, colors=['blue'], bestfit_colors=['pink']))

base_layout = cf.tools.get_base_layout(figs)
sp = cf.subplots( figs, shape=(3,2), base_layout=base_layout, vertical_spacing=.15, horizontal_spacing=.03,
                  specs=[[{'rowspan':2},{}],[None,{}],[{'colspan':2},None]],
                  subplot_titles=['四只股票收益率分布直方图', 'FUTU 和 NIO 散点图', '相关性矩阵热力图', 'DAO 拟合图'])
sp['layout'].update(showlegend=False)
cf.iplot(sp)
```

上面代码的难点在如何设定 specs。其参数值是二维列表，行列数由 shape 参数决定，本例是 3 行和 2 列。参数形式如 

    [[{'rowspan':2}, {}], [None, {}], [{'colspan':2}, None]]：

其中
    - {} 表示子图的占位符，即在此位置展示该子图，那么 {} 的个数要等于子图的个数，本例为 4
    - None 表示此位置不画图，那么 None 的个数要等于行数乘以列数再减去子图的个数，本例为 2

按照子图的顺序放入占位符得到，

    [ [{直方图},{散点图}], [None,{热力图}], [{拟合图},None] ]

对应的位置为，

    [ [{行1-2列1},{行1列1}], [None,{行2列2}], [{行3列1-2},None] ]
